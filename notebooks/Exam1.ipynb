## Problem 3

A bank has two types of accounts: checking & savings. All accounts can be identified by a unique account number and require a positive balance. Any checking or savings account allow withdrawal operation, however a savings account has a limit on number of withdrawal operations. A saving account earn interest that depends on an interest rate specified at the account opening. Checking accounts don't earn interest.

- Identify objects, their characteristics and responsibilities.
- Define classes for various accounts, use inheritance to reuse similar functionality for account types.
- Implement necessary methods so the below test driver program executed as expected and without errors.

This test driver program will help you to identify characteristics of account types, and provide hints for implementation of required responsibilities.

```c++
// Account Test Driver
//////////////////////
CheckingAccount ca{"ID1", 500};       // account number, money deposited
SavingsAccount  sa{"ID2", 1500, 1.0}; // account number, money deposited, interest rate

ca.withdraw(100);
ca.balance();       // Shows message: "Account[ID1]: $400"
sa.withdraw(100);
ca.balance();       // Shows message: "Account[ID2]: $1400"

sa.interest();
sa.balance();       // Shows message: "Account[ID2]: $1414"
//////////////////////
```

### Solution (without inheritance)

````c++
#include <iostream>
using namespace std;

class CheckingAccount {
    string ID;
    float sum;
public:
    CheckingAccount(string id, float s): ID{id}, sum{s} {}
    void balance() {
        cout << "Account[" << ID << "]: $" << sum << endl;
    }
    void withdraw(double wsum){
        if (sum-wsum > 0) {
            sum -= wsum;
        }
    }
};

class SavingsAccount {
    string ID;
    float sum;
    int limit;
    float rate;
public:
    SavingsAccount(string id, float s, float r): ID{id}, sum{s}, rate{r}, limit{3} {}
    void balance() {
        cout << "Account[" << ID << "]: $" << sum << endl;
    }
    void withdraw(double wsum){
        if (sum-wsum > 0 && limit > 0) {
            sum -= wsum;
            limit--;
        }
    }
    void interest(){ sum += sum*rate/100; }
};
````

### Solution (with inheritance)

````c++
#include <iostream>
using namespace std;

class Account {
protected:
    string ID;
    float sum;
public:
    Account() : ID{""}, sum{0.0} {}
    Account(string id, float s): ID{id}, sum{s} {}
    void balance() {
        cout << "Account[" << ID << "]: $" << sum << endl;
    }
};
class CheckingAccount: public Account {
    using Account::Account;
public:
    void withdraw(double wsum){
        if (sum-wsum > 0) {
            sum -= wsum;
        }
    }
};
class SavingsAccount: public Account {
    int limit;
    float rate;
public:
    SavingsAccount(string id, float s, float r): Account(id, s) {
        limit = 3; // only three operations allowed
        rate = r;
    }
    void withdraw(double wsum){
        if (sum-wsum > 0 && limit > 0) {
            sum -= wsum;
            limit--;
        }
    }
    void interest(){ sum += sum*rate/100; }
};
````
